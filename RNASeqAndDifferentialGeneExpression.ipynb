```python
#Run the quantification command
import subprocess

sras = ["SRR26704469","SRR26704467","SRR26704466","SRR26704462","SRR26704461","SRR26704460","SRR26704455",
               "SRR26704454","SRR26704453","SRR26704452","SRR26704451","SRR26704473",
               "SRR26704474","SRR26704465","SRR26704463","SRR26704464","SRR26704459","SRR26704468","SRR26704458",
               "SRR26704456","SRR26704457","SRR26704472","SRR26704471","SRR26704470"

]

index_folder = "RnaQuantification/transcripts_index/"
output_folder = "RnaQuantification/transcripts_quant"

for sra_id in sras:
    fastq1 = f"~/biol6150/ProjectSubmissions/Group19/Project56/AfterTrimming/{sra_id}_1.Trimmed.fastq"
    fastq2 = f"~/biol6150/ProjectSubmissions/Group19/Project56/AfterTrimming/{sra_id}_2.Trimmed.fastq"
    print(f"Running quantification for {sra_id}")
    sal_cmd=f"salmon quant -i {index_folder} -l A -1 {fastq1} -2 {fastq2} --validateMappings \
    -o {output_folder}/{sra_id} -p 6"

    subprocess.run(sal_cmd, shell=True)
```
